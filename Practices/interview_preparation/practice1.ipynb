```python
# simple triangle pattern

rows = 6

for num in range(rows):
    for i in range(num):
        print(num, end=" ") # print number
    # lines after each row to display pattern correctly
    print(" ")
```

```python
# creating function for simple traingle

def simpleTraingle(rows):
    for num in range(rows):
        for j in range(num):
            print(num, end=" ")
        print(" ")

simpleTraingle(6)
```

```python
# reverse simple triangle

rows = 5
b = 0
for i in range(rows, 0, -1):
    b += 1
    for j in range(1, i + 1):
        print(b, end = ' ')
    print('\r')
```

```python
def reverseTrangle(row):
    rows = row
    b = 0
    for i in range(rows, 0, -1):
        b += 1
        for j in range(1, i + 1):
            print(b, end = ' ')
        print('\r')

reverseTrangle(6)
```

```python
# Half pyramid pattern of numbers

rows = 5

for row in range(1, rows+1):
    for column in range(1, row + 1):
        print(column, end=' ')
    print("")
```

```python
# Inverted pyramid  of Descending order Numbers
rows = 5

for i in range(rows, 0, -1):
    num = i
    for j in range(0, i):
        print(num, end=' ')
    print("\r")
```

```python
# Pyramid of Natural Numbers less than 10

currentNumber = 1
stop = 2
rows = 3 # Rows you want in your pattern
for i in range(rows):
    for column in range(1, stop):
        print(currentNumber, end=' ')
        currentNumber += 1
    print(" ")
    stop += 2
```

```python
# Reverse pattern of digits from 10

start = 1
stop = 2
currentNumber = stop
for row in range(2, 6):
    for col in range(start, stop):
        currentNumber -= 1
        print(currentNumber, end=' ')
    print("")
    start = stop
    start += row
    currentNumber = stop
```

```python
# Unique pyramid pattern of digits

rows = 6

for i in range(1, rows + 1):
    for j in range(1, i - 1):
        print(j, end = " ")
    for j in range(i - 1, 0, -1):
        print(j, end=" ")
    print()
```

```python
# Even number pyramid pattern
rows = 5
lastEvenNumber = 2 * rows
evenNumber = lastEvenNumber
for i in range(1, rows + 1):
    evenNumber = lastEvenNumber
    for j in range(i):
        print(evenNumber, end=' ')
        evenNumber -= 2
    print("\r")
```

```python
# Pyramid of Horizontal Tables

rows = 7
for i in range(0, rows):
    for j in range(0, i + 1):
        print(i * j , end=' ')
    print()
```

```python
# Pyramid Pattern of Alternate Numbers

rows = 5
i = 1
while i <= rows:
    j = 1
    while j <= i:
        print((i * 2 -1), end = " ")
        j = j + 1
    i = i + 1
    print()
```

```python
# Mirroed Pyramid (Right-angled Triangle) Pattern of Numbers

rows = 6
for row in range(1, rows):
    num = 1
    for j in range(rows, 0, -1):
        if j > row:
            print(" ", end=' ')
        else:
            print(num, end=' ')
            num += 1
    print("")
```

```python
print("Print equilateral traingle Pyramid using stars \n")
size = 7
m = (2 * size) - 2
for i in range(0, size):
    for j in range(0, m):
        print(end=" ")
    m = m -1 # decrementing m after each loop
    for j in range(0, i + 1):
        # printing full Traingle pyramid using stars
        print("*", end=' ')
    print(" ")
```

```python
# Downward Traingle Pattern of Stars

rows = 5
k = 2 * rows - 2
for i in range(rows, -1, -1):
    for j in range(k, 0, -1):
        print(end=" ")
    k = k + 1
    for j in range(0, i + 1):
        print("*", end=" ")
    print("")
```

```python
# reverse string 

def reverseString(string):
    return string[::-1]

string = "Yash"
reverseString(string)
```

```python
# string is palindrome or not

def isPalindrome(string):
    if string[::-1] == string:
        print("The string is Palindrome")
    else:
        print("The string is not Palindrome")

string = "DFD"
isPalindrome(string)
```

```python
# fibonacci using recurssion in python

def fib(num):
    if (num <= 1):
        return num
    else:
        return (fib(num-1) + fib(num+1))
    
fib(2)
```
