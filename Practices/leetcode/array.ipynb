```python
# Two sum

nums = [2, 7, 9, 11]
target = 9

class solution:
    def twoSum(nums: list[int], target: int) -> list[int]:
        n = len(nums)
        for i in range(n-1):
            for j in range(i+1, n):
                if nums[i] + nums[j] == target:
                    return [i, j]
        return []
    

solution.twoSum(nums, target)
```
