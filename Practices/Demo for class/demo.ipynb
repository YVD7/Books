# files and Exceptions

```python
with open('pi_digits.txt', '+w') as f:
    contents = f.write("pi = 3.14")
    print(contents)
    f.close()
```

```python
with open('pi_digits.txt', 'r') as f:
    read = f.read()
    print(read)
```

```python
with open('pi_digits.txt', 'r') as f:
    read = f.read()
    print(read.rstrip())
```
