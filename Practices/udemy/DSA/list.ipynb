```python
myList = [1,2,3,4,5,6,7]
print(myList)
myList.insert(0,11)
myList.append(22)
print(myList)

newlist = [11,12,13,14,15,16,17]
myList.extend(newlist)
print(myList)
```

Searching in list

```python
myList = [10, 20, 30, 40, 50, 60, 70, 80, 90]

# in operator
target = 50
# if target in myList:
#     print(f"{target} is in the list")
# else:
#     print(f"{target} is not in the list")
```

```python
def linearSearch(p_list, p_target):
    for i, value in enumerate(p_list):
        if value == p_target:
            return i
    
    return -1

linearSearch(myList, target)
```
