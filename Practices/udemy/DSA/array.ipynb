```python
# create an array

import array

my_array = array.array('i', [1,2,3,4])
print(my_array)

# inserting
def insertArray(arr, index, value):
    print(arr.insert(index, value))
        
# traverse the array
def traverseArray(array):
    for i in array:
        print(i)

# accesiong the element of array
   
def accessElement(array, index): 
    if index > len(array):
        print("there is not an element this index")
    else:
        print(array[index])

# searching the element

def searchElement(array, index):
    for i in range(len(array)):
        if array[i] == index:
            return i
    return -1


# delete the element

def deleteEle(array, index):
    array.remove(index)
    print(array)

if __name__ == "__main__":

   insertion = insertArray(arr=my_array, index = 0, value=7)
   print(insertion)

   traverse =  traverseArray(my_array)
   print(traverse)

   access = accessElement(my_array, 4)
   print(access)

   searchEle = searchElement(my_array, 2)
   print(searchEle)

   deleteEle = deleteEle(my_array, 0)
   print(deleteEle)
```
