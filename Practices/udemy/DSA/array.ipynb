```python
# create an array

import array

my_array = array.array('i', [1,2,3,4])
print(my_array)

# inserting
def insertArray(array, index, value):
    print(f'array before insertion {array}')
    after_array = array.insert(index, value)
    print( f'after insertion {after_array} ')
        
# traverse the array
def traverseArray(array):
    for i in array:
        print(i)

# 

if __name__ == "__main__":

   insertion = insertArray(my_array, 0, 7)
   print(insertion)

   traverse =  traverseArray(my_array)
   print(traverse)

```
